**XOR using DNN**

```python
#packages

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# input and target XOR using DNN

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])


''' input and target AND using DNN
X=np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y=np.array([[0], [0], [0], [1]])
    input and target OR using DNN
X=np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y=np.array([[0], [1], [1], [1]]) '''
```

```python
# building the model

model = Sequential()

# Adding the dense layer for the hidden layer

model.add(Dense(8, input_dim=2, activation='relu'))

# Output layer

model.add(Dense(1, activation='sigmoid'))
```

```python
# compiling the model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fitting the input and target into the model

model.fit(X, y, epochs=1000, verbose=2)
```

```python
# calculating the loss and accuracy

loss, accuracy = model.evaluate(X, y)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
```

```python
# predicting the values

predictions = model.predict(X)
print("Predictions:")
for i in range(len(X)):
    print(f"Input: {X[i]}, Predicted Output: {predictions[i][0]:.4f}")
```
